```python
import os
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

%matplotlib inline
```

```python
models_dir = os.path.abspath(os.path.join('', '..', 'results/models'))
with open(os.path.join(models_dir, 'ComBat-GAM_rois_M'), 'rb') as f:
    ComBatGAM_rois_M = pickle.load(f)

with open(os.path.join(models_dir, 'ComBat-GAM_rois_F'), 'rb') as f:
    ComBatGAM_rois_F = pickle.load(f)
```

ROIs MALES

```python
sites = ['CHP', 'HSC', 'PNC']

site_CHP = stats.norm.rvs(size=10000, loc=ComBatGAM_rois_M['gamma_bar'][0], scale=np.sqrt(ComBatGAM_rois_M['t2'][0]))
sns.kdeplot(site_CHP, color='blue', label='CHP-prior')
sns.kdeplot(ComBatGAM_rois_M['gamma_hat'][0, :], color='blue', label='CHP-observed', linestyle='--').set_title('Location Effect in ROIs (Males)')
plt.show()

site_HSC = stats.norm.rvs(size=10000, loc=ComBatGAM_rois_M['gamma_bar'][1], scale=np.sqrt(ComBatGAM_rois_M['t2'][1]))
sns.kdeplot(site_HSC, color='blue', label='HSC-prior')
sns.kdeplot(ComBatGAM_rois_M['gamma_hat'][1, :], color='blue', label='HSC-observed', linestyle='--').set_title('Location Effect in ROIs (Males)')
plt.show()

site_PNC = stats.norm.rvs(size=10000, loc=ComBatGAM_rois_M['gamma_bar'][2], scale=np.sqrt(ComBatGAM_rois_M['t2'][2]))
sns.kdeplot(site_PNC, color='blue', label='PNC-prior')
sns.kdeplot(ComBatGAM_rois_M['gamma_hat'][2, :], color='blue', label='PNC-observed', linestyle='--').set_title('Location Effect in ROIs (Males)')
plt.show()
```

ROIs FEMALES

```python
site_CHP = stats.norm.rvs(size=10000, loc=ComBatGAM_rois_F['gamma_bar'][0], scale=np.sqrt(ComBatGAM_rois_F['t2'][0]))
sns.kdeplot(site_CHP, color='red', label='CHP-prior')
sns.kdeplot(ComBatGAM_rois_F['gamma_hat'][0, :], color='red', label='CHP-observed', linestyle='--').set_title('Location Effect in ROIs (Females)')
plt.show()

site_HSC = stats.norm.rvs(size=10000, loc=ComBatGAM_rois_F['gamma_bar'][1], scale=np.sqrt(ComBatGAM_rois_F['t2'][1]))
sns.kdeplot(site_HSC, color='red', label='HSC-prior')
sns.kdeplot(ComBatGAM_rois_F['gamma_hat'][1, :], color='red', label='HSC-observed', linestyle='--').set_title('Location Effect in ROIs (Females)')
plt.show()

site_PNC = stats.norm.rvs(size=10000, loc=ComBatGAM_rois_F['gamma_bar'][2], scale=np.sqrt(ComBatGAM_rois_F['t2'][2]))
sns.kdeplot(site_PNC, color='red', label='PNC-prior')
sns.kdeplot(ComBatGAM_rois_F['gamma_hat'][2, :], color='red', label='PNC-observed', linestyle='--').set_title('Location Effect in ROIs (Females)')
plt.show()
```
